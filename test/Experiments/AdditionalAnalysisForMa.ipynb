```python
import numpy as np
import matplotlib.pyplot as plt

heuristic = [0.92, 0.78, 0.69, 23.15, 22.57, 5.35, 6.53, 0.69, 0.44, 6.82, 7.68, 9.91, 7.11 ]
worst_case = [10.20, 9.54, 7.73, 258.47, 258.67, 95.44, 65.32, 9.01, 4.75, 74.37, 288.74, 129.72, 104.11]

m, b = np.polyfit(heuristic, worst_case, 1)
lin_approx = np.poly1d([m, b])

plt.plot(heuristic, worst_case, 'ro', heuristic, lin_approx(heuristic), '--b')
plt.title("Comparison of heuristic and worst case runtime")
plt.xlabel("Runtime heuristic (ms)")
plt.ylabel("Runtime worst case (ms)")
plt.legend(["Observed heuristic/worst case pairs", "Linear regression"])

plt.savefig("ComparionHeuristicWorstcase")
```
