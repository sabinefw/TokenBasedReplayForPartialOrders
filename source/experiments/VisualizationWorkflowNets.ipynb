```python
import pandas as pd
import pm4py
from pm4py.objects.conversion.log import converter as log_converter
from pm4py.objects.petri_net.importer import importer as pnml_importer
from pm4py.visualization.petri_net import visualizer as pn_visualizer
```

```python
viz_parameters = {
    "format": "png",
    "font_size": 10,
    "place_size": 15,
    "arc_width": 0.5,
    "node_sep": 0.3,
    "rank_sep": 0.3,
    "rankdir": "LR"  # Change to "TB", "BT", "LR", or "RL"
}
```

```python
repair_net, repair_initial_marking, repair_final_marking = pnml_importer.apply("data/nets/repair_net.pnml")
repair_net_viz = pn_visualizer.apply(repair_net, repair_initial_marking, repair_final_marking, parameters=viz_parameters)
pn_visualizer.view(repair_net_viz)
```

```python
pn_visualizer.save(repair_net_viz, "visualization/repair.png")
```
